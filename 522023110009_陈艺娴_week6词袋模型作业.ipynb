```python
# 导入数据
import pandas as pd
```

###最开始代码报错，错误原因显示在尝试解码CSV文件时出现了问题。 'utf-8' 编解码器无法解码位置0处的字节0xd0，因此要重新设置赋值'gbk'编码格式

```python
text_ana = pd.read_csv('C:/Users/20693/Desktop/智慧养老政策.csv', encoding='gbk',index_col = 0)
# text_ana = text_ana.drop(['Unnamed: 0'], axis=1)
text_ana
```

```python
# 分析语料
text = list(text_ana['标题'])
text
```

###基础的标题文本数据已经获取

```python
import jieba
def word_segmentation(text):
    words = jieba.lcut(text)
    return ' '.join(words)
```

```python
text_ana['标题'] = text_ana['标题'].iloc[:100]\
                          .astype(str).apply(word_segmentation)
text_ana['标题'].head()
```

```python
text = list(text_ana['标题'].dropna())
text
```

###至此，所有的分词已经完成，基础的文本数据已经全部处理结束。接下来就要开始构建词袋模型

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(text)

print('词袋特征矩阵的形状:', X.shape)
```

```python
print('特征词列表:', vectorizer.get_feature_names())
```

###根据错误提示，'CountVectorizer'对象没有名为 'get_feature_names_out' 的属性或方法，提示应该修改为get_feature_names()'，去掉out

```python
list(X.toarray())
```

###这一步已经将文本数据，转化为计算机代码语言

```python
feature_words = vectorizer.get_feature_names()
word_freq = dict(zip(feature_words, X.sum(axis=0).A1))
len(word_freq)
```

```python
sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
```

###这一步把基础的文本转化成计算机程序里可以读取检索技数的形式，开始对具体的文本数量进行归类和分析

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
if not isinstance(text_ana, str):
    text_ana = str(text_ana)
```

```python
font_path = "C:/Windows/Fonts/simsun.ttc"
```

```python
wc = WordCloud(font_path=font_path).generate(text_ana)

# 显示词云图
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# 根据词袋模型分析结果绘制词云图
wordcloud = WordCloud(
    font_path='C:/Windows/Fonts/simsun.ttf',  # 指定中文字体文件路径
    background_color='white',  # 设置背景颜色
    width=1200,
    height=800,
    max_words=200,  # 设置最大显示的词数
    max_font_size=100  # 设置字体最大值
).generate_from_frequencies(word_freq)

# 显示词云图
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
